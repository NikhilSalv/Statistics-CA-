### Stationary test

```python
# In order to check the data is standard or not, we will apply a test called dicky fuller test
# To apply dicky fuller test, we use Stat model called stattools. (adfuller)
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# Dicky fuller test is a robust test to determine whether data is stationary or not.
# It is a hypothesis test ,in which null hypothesis is: The data is not stationary
# whereas alternate hypothesis is, data is standard

# The adfuller package gives is 5 components, which are: 
# ADF test statistics, p value, number of lags used, and number of observations used.

# P value is significant as it decides whether the data is standard or not
# If p value is below 5% ,that is, if it is less than critical value, we reject the null hypothesis.
# We are going to define a function to check whether the data is stationary or not.
```

```python
def adfuller_test(data):
    result = adfuller(data)
    labels = ["ADF Test statistics","p value", "number of lags used", "number of observations used"]
    for r, l in zip(result,labels):
        print(str(l) + " : " + str(r))
    if result[1] <= 0.05:
        print("Reject null, data is stationary")
    else:
        print("Fail to reject null, data is not stationary")
```

```python
adfuller_test(df["Exp_Smooth"])
```

```python
# This says that we need to make the data stationary.
# We are going to make data stationary by differencing
```

### Differencing

```python
# Since, we have observed in the previous visualisation that data is seasonal, instead of doing just single shift difference,
# We do seasonal difference, that means, we will shift the data by 12, because it is a monthly data (and 1 year means 12 months)
```

```python
df["Exp_Seas_Diff"] = df["Exp_Smooth"] - df["Exp_Smooth"].shift(1)
```

```python
# df["Seasonal_diff"] = df["Registrations"] - df["Registrations"].shift(12)
```

```python
# The p value is less than 5% , hence, we reject the null hypothesis. 
# This means, the data is stationary.
```

#### Auto Regressive Model

> We are going to identify an AR model by using PACF (Partial Auto Correlation)

For an AR model, the theoretical PACF "shuts off" past the order of the model.

For MA model, the theoretical PACF does not shut off, but tapers towards 0 in some manner.

Hence, for MA model, a clearer pattern is ACF.

ARIMA model has 3 components: p,d and q

p : AR lags

d : Differencing

q : MA lag

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot_pacf(df["Exp_Seas_Diff"])
```

From the above model, we can observe that after 2, the PACF shuts down. 

We will testify the value of p with 3 and 4, and analyse which gives us the best results.

```python
plot_acf(df["Exp_Seas_Diff"])
```

From the above analysis, we can observe that, the exponential graph goes under critical limit after 2. 

Hence, for our MA model, we will testify values 2 ,3, and 4. (Q)

For Differencing, we will use value 1.
